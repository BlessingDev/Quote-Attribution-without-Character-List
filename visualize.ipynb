```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
import numpy as np
```

```python
x = np.arange(1, 20)
y_sq = x ** 2
y_li = x
y_e = np.exp(x)
y_xlogx = x * np.log(x)

plt.plot(x, y_sq, label="x**2")
plt.plot(x, y_li, label="x")
#plt.plot(x, y_e, label="e**x")
plt.plot(x, y_xlogx, label="x*log(x)")
plt.legend()

plt.show()
```
